```python
df = sns.load_dataset("penguins")
df.head()
```

```python
# Check for missing values
df.isnull().sum()

# Drop rows with missing values
df.dropna(inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

# Label encoding for 'species', 'island', and 'sex'
le = LabelEncoder()
df['species'] = le.fit_transform(df['species'])
df['island'] = le.fit_transform(df['island'])
df['sex'] = df['sex'].map({'MALE': 1, 'FEMALE': 0})  # Male = 1, Female = 0
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_features = scaler.fit_transform(df.drop('species', axis=1))

# Convert back to DataFrame
df_scaled = pd.DataFrame(scaled_features, columns=df.drop('species', axis=1).columns)
df_scaled['species'] = df['species']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Pairplot to visualize relationships between features
sns.pairplot(df, hue='species', diag_kind='kde')
plt.show()
```

```python
corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier

# Train a Random Forest model to get feature importance
X = df.drop('species', axis=1)
y = df['species']

rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X, y)

# Feature importance
importance = pd.Series(rf.feature_importances_, index=X.columns)
importance.sort_values(ascending=False, inplace=True)
importance.plot(kind='bar')
plt.title('Feature Importance')
plt.show()
```

```python
# Check for missing values in X_train and X_test
print(X_train.isnull().sum())
print(X_test.isnull().sum())
```

```python
from sklearn.impute import SimpleImputer

# Create an imputer object to fill missing values with the mean
imputer = SimpleImputer(strategy='mean')

# Fit the imputer on the training data and transform it
X_train_imputed = imputer.fit_transform(X_train)

# Transform the test data using the fitted imputer
X_test_imputed = imputer.transform(X_test)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Initialize the SVM model
svm = SVC(kernel='linear')

# Train the model with the imputed data
svm.fit(X_train_imputed, y_train)

# Predict on the test set
y_pred_svm = svm.predict(X_test_imputed)

# Evaluate the accuracy
print(f"SVM Accuracy: {accuracy_score(y_test, y_pred_svm)}")
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Create a pipeline with imputation and SVM
pipeline = make_pipeline(
    SimpleImputer(strategy='mean'),  # Impute missing values
    SVC(kernel='linear')             # Train the SVM model
)

# Fit the model
pipeline.fit(X_train, y_train)

# Predict on the test set
y_pred_svm = pipeline.predict(X_test)

# Evaluate the accuracy
print(f"SVM Accuracy: {accuracy_score(y_test, y_pred_svm)}")
```

```python
import xgboost as xgb

# Train XGBoost model
xg_model = xgb.XGBClassifier(use_label_encoder=False, eval_metric='mlogloss')
xg_model.fit(X_train, y_train)

# Predict on test set
y_pred_xgb = xg_model.predict(X_test)

# Evaluate accuracy
print(f"XGBoost Accuracy: {accuracy_score(y_test, y_pred_xgb)}")
```

```python
from sklearn.model_selection import GridSearchCV

# Hyperparameter tuning for RandomForest
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5, 10]
}

grid_search = GridSearchCV(RandomForestClassifier(), param_grid, cv=5, n_jobs=-1, verbose=2)
grid_search.fit(X_train, y_train)

# Best parameters and score
print(f"Best Parameters: {grid_search.best_params_}")
print(f"Best Score: {grid_search.best_score_}")
```
