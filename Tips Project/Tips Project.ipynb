```python
# Data Cleaning
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df['sex']=df['sex'].map({'Female':0,'Male':1})
df['smoker']=df['smoker'].map({'No':0,'Yes':1})
```

```python
# Boxplot to check the distribution of tips by day
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='day', y='tip', data=df,palette='Accent')
plt.title('Distribution of Tips by Day')
plt.show()
```

```python
# Plot a scatter plot of total bill vs tip
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='total_bill', y='tip',hue='sex', data=df,palette='BuPu')
plt.title('Total Bill vs Tip')
plt.show()
```

```python
# Count plot for the sex variable
```

```python
plt.figure(figsize=(5, 5))
sns.countplot(x='sex', data=df,linewidth=0.25,width=0.25,palette=['r','grey'])
plt.title('Count of Diners by Sex')
plt.show()
```

```python
# Question 1: What is the average tip for each day of the week?
```

```python
avg_tip_by_day = df.groupby('day')['tip'].mean()
print(avg_tip_by_day)
```

```python
#Question 2: What is the relationship between the total bill and the tip?
```

```python
correlation = df['total_bill'].corr(df['tip'])
print(f"The correlation between total bill and tip is: {correlation:.2f}")
```

```python
#Question 3: What is the average bill amount by gender?
```

```python
avg_bill_by_sex = df.groupby('sex')['total_bill'].mean()
print(avg_bill_by_sex)
```

```python
#Question 4: How does the tip amount vary based on the time of day (Lunch or Dinner)?
```

```python
avg_tip_by_time = df.groupby('time')['tip'].mean()
print(avg_tip_by_time)
```

```python
#Question 5: What is the proportion of tips given for different sizes of parties?
```

```python
tip_by_size = df.groupby('size')['tip'].mean()
print(tip_by_size)
```

```python
#Question 6: How do tips vary based on the restaurant day of the week?
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='day', y='tip', data=df,palette='BuPu')
plt.title('Tips by Day')
plt.show()
```

```python
#Question 7: Is there a noticeable difference in tips based on whether customers are male or female?
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='sex', y='tip', data=df)
plt.title('Tip Distribution by Sex')
plt.show()
```

```python
#Question 8: Does the total bill amount increase as the party size grows?
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='size', y='total_bill', data=df)
plt.title('Total Bill vs Party Size')
plt.show()
```

```python
# Question 9: What percentage of customers leave a tip less than $2?
```

```python
tip_less_than_2 = df[df['tip'] < 2].shape[0] / df.shape[0] * 100
print(f"Percentage of customers leaving a tip less than $2: {tip_less_than_2:.2f}%")
```

```python
# Question 10: What are the most common sizes of parties?
```

```python
party_sizes = df['size'].value_counts()
print(party_sizes)
```

```python
#  Custom Colors Using Hex Codes
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='sex', data=df, palette=['#FF6347', '#4682B4'],saturation=75)
plt.title('Count of Diners by Sex')
plt.show()
```

```python
# Using hue for Grouping and Coloring
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='sex', data=df, hue='smoker', palette='coolwarm')  
plt.title('Count of Diners by Sex and Smoking Status')
plt.show()
```

```python

custom_palette = sns.color_palette(["#FF6347", "#4682B4", "#32CD32", "#FFD700"])
# Count plot with a custom palette
plt.figure(figsize=(8, 6))
sns.countplot(x='day', data=df, palette=custom_palette)  # Multiple colors for days
plt.title('Count of Diners by Day')
plt.show()
```

```python
# Using hue and Custom Palette for Multiple Colors
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='sex', data=df, hue='smoker', palette=['#FF6347', '#4682B4'])
plt.title('Count of Diners by Sex and Smoking Status')
plt.show()
```

```python
 # Using edgecolor to Set the Line Color
```

```python
# Count plot with a specific line color for the edges of the bars
plt.figure(figsize=(8, 6))
sns.countplot(x='sex', data=df, color='red', edgecolor='grey',linewidth=1.65,width=0.25)  # Set line color to black
plt.title('Count of Diners by Sex')
plt.show()
```

```python
# Using edgecolor with Custom Colors
```

```python
plt.figure(figsize=(4, 4))
sns.countplot(x='sex', data=df, color='grey', edgecolor='#FF6347',width=0.25)
plt.title('Count of Diners by Sex')
plt.show()
```

```python
# Using edgecolor with a Custom Palette and Different Line Colors
```

```python
plt.figure(figsize=(4, 6))
sns.countplot(x='sex', data=df, palette=['#FF6347', '#4682B4'], edgecolor='black')
plt.title('Count of Diners by Sex')
plt.show()
```
