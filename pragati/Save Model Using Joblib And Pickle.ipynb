```python
import pandas as pd
import numpy as np
```

```python

train_df = pd.read_csv(r"C:\Users\Admin\Downloads\homeprices.xls")
train_df.head()
```

```python
import math
bedrooms_mean = math.floor(train_df.bedrooms.mean())
train_df.bedrooms = train_df.bedrooms.fillna(bedrooms_mean)
train_df
```

```python
train_X = train_df.drop(['price'], axis=1)
train_Y = train_df.price
train_X,train_Y
```

```python
# Train model
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(train_X, train_Y)
```

```python
model
```

```python
model.coef_
```

```python
model.intercept_
```

```python
#predict home price
price_predicated = model.predict([[3000, 3, 40]])
price_predicated_manually = 3000*model.coef_[0] + 3 * model.coef_[1] + 40 * model.coef_[2] + model.intercept_
price_predicated, price_predicated_manually
```

```python
price_predicated
```

```python
# Score
model.score(train_X, train_Y)
```

```python
import pickle
with open('model_pickle', 'wb') as f:
    pickle.dump(model, f)
```

```python
# import trained model using pickle
with open('model_pickle', 'rb') as f:
    pickle_model = pickle.load(f)
price_predicated_pickle = pickle_model.predict([[3000, 3, 40]])
price_predicated_pickle
```
